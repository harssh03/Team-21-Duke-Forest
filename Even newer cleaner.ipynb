## Ant observations addition

```python
def add_ant_obs(df):
    herp_obs_index = df.columns.get_loc('Herp_obs') + 1
    df.insert(herp_obs_index, 'Ant_obs', 'No')
    
    sp_columns = [f'SP_{i}' for i in range(1, 16)]

    # Update 'Ant_obs' column based on the presence of 'ant' in SP columns
    for index, row in df.iterrows():
        if any('ant' in str(row[sp_col]).lower() for sp_col in sp_columns):
            df.at[index, 'Ant_obs'] = 'Yes'

    return df
```

```python
def wide_to_long(df):
    new_rows = []
    for index, row in df.iterrows():
        herp_obs = row['Herp_obs']
        ant_obs = row['Ant_obs']
        if herp_obs == 'No' and ant_obs == 'No':
            # Include only the first SP observation
            new_row = row.copy()
            new_row['SP'] = row['SP_1']
            new_row['SP_Cover_ID'] = row['SP_1_Cover_ID']
            new_row['SP_Total'] = row['SP_1_Total']
            new_row['SP_Photo'] = row['SP_1_Photo']
            new_row['Dead_animal'] = np.nan
            new_row['Dead_detail'] = np.nan
            new_row['Other_animal'] = np.nan
            new_row['Other_detail'] = np.nan
            new_rows.append(new_row)
        else:
            # Include all non-NaN SP observations
            for i in range(1, 16):
                sp = row[f'SP_{i}']
                if pd.isna(sp):
                    continue
                new_row = row.copy()
                new_row['SP'] = row[f'SP_{i}']
                new_row['SP_Cover_ID'] = row[f'SP_{i}_Cover_ID']
                new_row['SP_Total'] = row[f'SP_{i}_Total']
                new_row['SP_Photo'] = row[f'SP_{i}_Photo']
                new_row['Dead_animal'] = np.nan  # Placeholder, update based on your logic
                new_row['Dead_detail'] = np.nan  # Placeholder, update based on your logic
                new_row['Other_animal'] = np.nan  # Placeholder, update based on your logic
                new_row['Other_detail'] = np.nan  # Placeholder, update based on your logic
                new_rows.append(new_row)

    # Create a new DataFrame from the new rows
    new_df = pd.DataFrame(new_rows)
    
    return new_df
```
